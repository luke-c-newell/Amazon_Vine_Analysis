```python
# Import dependencies
import pandas as pd
```

```python
# Create variable for csv file
vine_data = "vine_table.csv"
```

```python
# Import csv file as a dataframe and show data
vine_table_df = pd.read_csv(vine_data)
vine_table_df.head()
```

```python
# Create a new dataframe to show only reviewers where total_votes > 20
total_votes_df = vine_table_df.loc[vine_table_df['total_votes'] >= 20]
total_votes_df.head()
```

```python
# Filter the DataFrame to retrieve rows where helpful_votes / total_votes > 50%
useful_votes_df = total_votes_df.loc[(total_votes_df['helpful_votes'] / total_votes_df['total_votes']) >= 0.5]
useful_votes_df.head(30)
```

```python
# Create a DataFrame where all the reviews are part of the Vine program (paid reviews)
vine_reviews = useful_votes_df.loc[useful_votes_df['vine'] == 'Y']
vine_reviews.head()
```

```python
# Create a DataFrame where all the reviews are NOT part of the Vine program (unpaid reviews)
non_vine_reviews = useful_votes_df.loc[useful_votes_df['vine'] == 'N']
non_vine_reviews.head()
```

```python
# Determine total number of paid reviews
len(vine_reviews)
```

```python
# Determine total number of unpaid reviews
len(non_vine_reviews)
```

```python
# Calculate the number of five star reviews for non vine reviews
five_star_non_vine_reviews = non_vine_reviews.loc[non_vine_reviews['star_rating'] == 5]
len(five_star_non_vine_reviews)
```

```python
# Calculate the number of five star reviews for vine reviews
five_star_vine_reviews = vine_reviews.loc[vine_reviews['star_rating'] == 5]
len(five_star_vine_reviews)
```

```python
# Calculate the percentage of five star reviews for non vine reviews
len(five_star_non_vine_reviews) / len(non_vine_reviews) * 100
```

```python
# Calculate the percentage of five star reviews for vine reviews
len(five_star_vine_reviews) / len(vine_reviews) * 100
```
